**Introduction to Matplotlib**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Basic Plotting

# Sample data
x = np.linspace(0, 10, 100)  # Generate 100 evenly spaced points between 0 and 10
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

The plot() function

* The plot() function of matplotlib is a versatile plotting function that can create various types of plots, including line plots, scatter plots, and step plots.
* Arguments:-
  - x: The x-coordinates of the data points.
  - y: The y-coordinates of the data points.
  - color: The color of the line or markers
  - linestyle: The style of the line connecting the points (solid/dashed/dot-dashed/dotted)
  - linewidth: The width of the line
  - marker: The shape of the data point markers (circle/square/star/diamond etc.)
  - markersize: The size of the marker
  - label: label for the plot
  - alpha: The transparency level of the plot
  - markerfacecolor: The color of the marker's face (interior)
  - markeredgecolor: The color of the marker's edge (border)

**Customizations possible:-**

```python
# Stacked Area Chart
months = range(1, 13)

# Mushroom yield data for four years
m1 = [14, 13, 10, 15, 17, 15, 15, 13, 12, 10, 14, 11]
m2 = [16, 12, 11, 14, 16, 14, 16, 12, 13, 11, 15, 12]
m3 = [15, 14, 9, 16, 18, 16, 14, 14, 11, 9, 13, 10]
m4 = [13, 15, 12, 13, 15, 13, 17, 11, 14, 12, 16, 13]

plt.stackplot(months,m1, m2, m3, m4, labels=['2024','2023','2022','2021'])
plt.legend(loc='upper left')

# Labels and Title
plt.xlabel('Month')
plt.ylabel('Mushroom Yield')
plt.title('Mushroom Yield Over 4 Years')

# Set x-axis ticks to show month names (optional)
plt.xticks(months,
           ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
            'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Improve layout
plt.tight_layout()

# Show Plot
plt.show()
```

```python
# 100% Stacked Area Chart
data = pd.DataFrame({  'group_A':[1,4,6,8,9],
                     'group_B':[2,24,7,10,12],
                       'group_C':[2,8,5,10,6], }, index=range(1,6))

# We need to transform the data from raw data to percentage (fraction)
data_perc = data.divide(data.sum(axis=1), axis=0)
print(data_perc)

# Make the plot
plt.stackplot(range(1,6),  data_perc["group_A"],  data_perc["group_B"],  data_perc["group_C"], labels=['A','B','C'])
plt.legend(loc='upper left')
plt.margins(0,0)
plt.title('100 % stacked area chart')
plt.show()
```

Boxplot mainly provides a graphical summary of a data set with features such as minimum, first quartile, median, third quartile, and maximum.


*   the box's beginning - first quartile
*   the box's end - third quartile
*   line inside the box - median value
*   x-axis of a box plot represents the data to be plotted
*   y-axis represents the frequency distribution
*   whiskers - lines extending from the box to the minimum and maximum values within 1.5 * IQR from the lower and upper quartiles, respectively
*   outliers - data points outside the whiskers



*   IQR=Q3Q1
*   Lower Whisker: 11.5IQR
*   Upper Whisker: Q3+1.5IQR

```python
# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)
# np.random.normal(loc=100, scale=20, size=200)
# generates 200 random data points following a normal (Gaussian) distribution
# with a mean of 100 and a standard deviation of 20.

fig = plt.figure(figsize =(10, 7))
# 10 in width, 7 in height

# Creating plot
plt.boxplot(data)

# show plot
plt.show()
```

A bubble plot is similar to a scatter plot but adds an extra dimension by varying the size of the markers.


*   X-axis: First variable.
*   Y-axis: Second variable.
*   Bubble Size: Represents the magnitude of the third variable.
*   Bubble Color (Optional): Represents a fourth variable, adding more depth to the visualization.



```python
# Seed for reproducibility
np.random.seed(42)

# Number of data points
num_countries = 30

# Generate sample data
countries = [f"Country {i+1}" for i in range(num_countries)]
gdp_per_capita = np.random.uniform(1000, 50000, num_countries)  # in USD
life_expectancy = np.random.uniform(50, 85, num_countries)      # in years
population = np.random.uniform(1e6, 1e8, num_countries)        # in people


# Create a figure and axis
plt.figure(figsize=(12, 8))

# Create the bubble plot
scatter = plt.scatter(
    gdp_per_capita,        # X-axis: GDP per capita
    life_expectancy,       # Y-axis: Life Expectancy
    s=population / 1e6,    # Bubble size: Population (scaled)
    alpha=0.6,             # Transparency
    c=gdp_per_capita,      # Bubble color: GDP per capita
    cmap='viridis',        # Color map
    edgecolors='w',        # White edge for better visibility
    linewidth=0.5
)

# Add color bar
cbar = plt.colorbar(scatter)
cbar.set_label('GDP per Capita (USD)', fontsize=12)

# Add labels and title
plt.xlabel('GDP per Capita (USD)', fontsize=14)
plt.ylabel('Life Expectancy (Years)', fontsize=14)
plt.title('Bubble Plot: GDP per Capita vs Life Expectancy', fontsize=16)

# Add grid
plt.grid(True, linestyle='--', alpha=0.5)

# Show plot
plt.show()
```

10.   Word Clouds



Used for representing text data, the size of words are dependent on the occurrence frequency (texts importance in the context), therefore it comes in handy for Natural Language Processing machine learning projects or text analysis.

```python
import requests
import string
from bs4 import BeautifulSoup
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize the text
    tokens = text.split()

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    cleaned_tokens = [word for word in tokens if word not in stop_words]

    # Join tokens back to a single string
    cleaned_text = ' '.join(cleaned_tokens)

    return cleaned_text
```

```python
# get URL
netflix = requests.get("https://en.wikipedia.org/wiki/Netflix")

# store content into variable
netflix_text = BeautifulSoup(netflix.content, 'html.parser').text

# use function to clean text
netflix_text = clean_text(netflix_text)
print(netflix_text)
```

```python
# Plot the wordcloud with the mask applied
wc = WordCloud(background_color='black', contour_width=2, contour_color='black', colormap = 'YlOrRd', stopwords = ['Netflix']).generate(netflix_text)
plt.figure(figsize=[16,16])
plt.tight_layout()
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
import seaborn as sns
```

```python
# Sample Data
np.random.seed(0)
data = pd.DataFrame({
    'x': np.arange(1, 11),
    'y': np.random.randint(1, 100, size=10),
    'category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A']
})

data.head()
```

```python
# Create a figure to hold all plots
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Line plot
sns.lineplot(data=data, x='x', y='y', ax=axs[0, 0])
axs[0, 0].set_title('Line Plot')

# Area plot
axs[0, 1].fill_between(data['x'], data['y'], color="skyblue", alpha=0.4)
axs[0, 1].plot(data['x'], data['y'], color="Slateblue", alpha=0.6)
axs[0, 1].set_title('Area Plot')

# Bar plot
sns.barplot(data=data, x='x', y='y', ax=axs[1, 0], palette='viridis')
axs[1, 0].set_title('Bar Plot')

# Histogram
sns.histplot(data['y'], bins=10, kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Histogram')

# Pie chart
axs[2, 0].pie(data['y'], labels=data['category'], autopct='%1.1f%%', startangle=90)
axs[2, 0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[2, 0].set_title('Pie Chart')

# Scatter plot
sns.scatterplot(data=data, x='x', y='y', hue='category', style='category', ax=axs[2, 1], s=100)
axs[2, 1].set_title('Scatter Plot')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

Regression plot

```python
# load the dataset
dataset = sns.load_dataset('tips')

dataset.head()
```

Regression plots in seaborn can be implemented with the help of the lmplot() function.
lmplot() can be understood as a function that basically creates a linear model plot. It creates a scatter plot with a linear fit on top of it.

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = dataset, hue ='sex',
           markers =['o', 'v'], scatter_kws ={'s':100},
           palette ='plasma')
```



*   The line of best fit (regression line) represents the relationship between total_bill and tip
*   A positive slope indicates that, on average, as the total bill increases, the tip also increases.
*   The steepness of the slope can indicate how much more the tip tends to increase with a higher bill.

```python
import folium

tiles_map = folium.Map(location=[40.7128, -74.0060], tiles=None)

folium.TileLayer(
    'Stamen Terrain',
    name='Stamen Terrain',
    attr='Stamen Design, OpenStreetMap'
).add_to(tiles_map)

folium.TileLayer(
    'Stamen Toner',
    name='Stamen Toner',
    attr='Stamen Design, OpenStreetMap'
).add_to(tiles_map)

folium.TileLayer(
    'Stamen Watercolor',
    name='Stamen Watercolor',
    attr='Stamen Design, OpenStreetMap'
).add_to(tiles_map)

folium.TileLayer(
    'cartodbpositron',
    name='CartoDB Positron',
    attr='CartoDB'
).add_to(tiles_map)

folium.TileLayer(
    'cartodbdark_matter',
    name='CartoDB Dark Matter',
    attr='CartoDB'
).add_to(tiles_map)

folium.TileLayer('OpenStreetMap', name='OpenStreetMap').add_to(tiles_map)
tiles_map.add_child(folium.LayerControl())

tiles_map
```

Heatmap:-

A heat map (or heatmap) is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions.

```python
from folium.plugins import HeatMap

# create a map object
mapObj = folium.Map([24.2170111233401, 81.0791015625000], zoom_start=6)

# data for heatmap.
# each list item should be in the format [lat, long, value]
data = [
    [24.399, 80.142, 0.78],
    [22.252, 80.885, 0.68],
    [24.311, 80.543, 0.58],
    [23.195, 82.994, 0.46],
    [23.431, 80.427, 0.76],
    [26.363, 81.791, 1.81],
    [22.942, 83.257, 0.75],
    [23.751, 79.995, 0.16],
    [23.215, 81.004, 0.64],
    [24.541, 79.889, 0.55]
]

# create heatmap from the data and add to map
HeatMap(data).add_to(mapObj)

mapObj
# areas with higher cumulative value will appear "hotter" (e.g., red)
# areas with lower values will appear "cooler"
```
